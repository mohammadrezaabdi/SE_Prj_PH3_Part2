```python
#[0,1,7,2,3,8]
# arr = [5]
# x = 5
# expected = 0
# Possible
# -----------
# [0,1,7,2,5,7,6,8]
# arr = [1]
# x = 4
# expected = -1
# Impossible
# -----------
# [0,1,7,2,4,7,6,8]
# arr = [5]
# x = 4
# expected = -1
# Impossible
def find_first_in_sorted(arr, x):
    lo = 0
    hi = len(arr)

    while lo <= hi:
        mid = (lo + hi) // 2

        if x == arr[mid] and (mid == 0 or x != arr[mid - 1]):
            return mid

        elif x <= arr[mid]:
            hi = mid

        else:
            lo = mid + 1

    return -1

print(find_first_in_sorted([5], 4))
"""
Fancy Binary Search
fancy-binsearch

Input:
    arr: A sorted list of ints
    x: A value to find

Output:
    The lowest index i such that arr[i] == x, or -1 if x not in arr
"""
```
